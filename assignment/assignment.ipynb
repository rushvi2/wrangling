1. This paper is about the most efficient method to clean datasets. It talks about how to format rows, columns, and tables in order to make the data easier to manipulate and visualize.
2. The tidy data standard is meant to make data exploration and analysis easier and simplify the development of data analysis tools. The tools and cleaning methods that are currently being used are hard to understand for most statisticians. This method is trying to standardize the cleaning of datasets and allows advice for how the datasets should be structured and how they can work with other data analysis tools.
3. "Like families, tidy datasets are all alike but every messy dataset is messy in its own way." This sentence talks about how with tidy datasets, there is a common theme among the data and it is known where what is. There are no ambiguities in the dataset. With messy datasets, there could be a different problem that needs to be fixed each time. There is no set problem that a dataset could have which makes fixing these errors hard. "For a given dataset, it’s usually easy to figure out what are observations and what are variables, but it is surprisingly difficult to precisely define variables and observations in general." This sentence means that some datasets might have observations that are not intuitive unless it is interpreted in the context of the entire dataset. Misjudging what a value might mean might mess up any additional cleaning and analysis that happens further along.
4. Values are what make up a dataset and are usually numbers or strings. Each value belongs to a variable and an observation. A variable contains all values that are measured across units with the same attribute. An observation contains all values measured along the same unit across attributes.
5. "Tidy Data" is defined as a dataset where each variable forms a column, each observation forms a row, and each type of observational unit forms a table.
6. The 5 most common problems are that column headers are values and not variable names, multiple variables are stored in one column, variables are stored in both rows and columns, multiple types of observational units are stored in the same table, and a single observational unit is stored in multiple tables. The data in table 4 are messy because the dataset contains values as column headers and should contain an additional variable named income. Melting a dataset means to turn the columns into rows.
7. Table 11 is messy because it has variables stored in both the rows and columns. Table 12 is molten because the dates, year variable and month variable turn into just date. The table then becomes tidy by eliminating the element column since it contains variable names. The tidy table contains just values.
8. The chicken-and-egg problem is that if the tidy data is only as useful as the tools that work with it, then these tools will not be able to be separated from tidy data. Wickham hopes that others build upon the framework to develop better tools for data cleaning and analysis in a broader sense.

```python
df = pd.read_csv('./data/sharks.csv', low_memory=False)
type = df['Type']
type = type.replace(['Sea Disaster', 'Boat','Boating','Boatomg'],'Watercraft')
type = type.replace(['Invalid', 'Questionable','Unconfirmed','Unverified','Under investigation'],np.nan)
df['Type'] = type
df['Type'].value_counts()
del type
```

```python
df = pd.read_csv('/content/wrangling/data/October 2017 Cohort_Virginia Pretrial Data Project_Deidentified FINAL Update_10272021.csv', low_memory=False)
df.head()
```

```python
df['WhetherDefendantWasReleasedPretrial'] = df['WhetherDefendantWasReleasedPretrial'].replace(9,np.nan)
print(df['WhetherDefendantWasReleasedPretrial'].value_counts(),'\n')
```

```python
imp = df['ImposedSentenceAllChargeInContactEvent']
sentence = df['SentenceTypeAllChargesAtConvictionInContactEvent']
imp = pd.to_numeric(imp,errors='coerce')  #change to numeric
imp = imp.mask( sentence == 4, 0) #replace type 4 with 0
imp = imp.mask( sentence == 9, np.nan) #replace 9 with nan
print( pd.crosstab(imp.isnull(), sentence), '\n')

df['ImposedSentenceAllChargeInContactEvent'] = imp
del imp, sentence
```
